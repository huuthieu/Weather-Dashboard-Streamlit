```python
def split_m1(x):
    x = str(x)
    if len(x) == 7:
        return  x[0]
    else:
        return str(int(x[:2]))

def split_d1(x):
    x = str(x)
    if len(x) == 7:
        return str(int(x[1:3]))
    else:
        return str(int(x[2:4]))

def split_y1(x):
    x = str(x)
    if len(x) == 7:
        return x[3:]
    else:
        return x[4:]
```

```python
def split(x):
    x = str(x)
    if len(x) == 7:
        return x[0]
    else:
        return x[:2]

def split_m(x):
    x = str(x)
    if len(x) == 7:
        return str(int(x[1:3]))
    else:
        return str(int(x[2:4]))

def split_y(x):
    x = str(x)
    if len(x) == 7:
        return x[3:]
    else:
        return x[4:]
```

```python
def process_data_freeme_sung(year):
    data = pd.read_csv(f"../data_sung/freemeteo{year}.csv")
    data["Day"] = data["Ngày"].apply(split)
    data["Month"] = data["Ngày"].apply(split_m)
    data["year"] = data["Ngày"].apply(split_y)
    data = data.drop(columns=["Ngày","Mô tả","Ngày thu thập","url"])
    
    return data.rename(columns={"Thời gian":"Time"})
```

```python
def process_data_world_sung(year):
    data = pd.read_csv(f"../data_sung/worldweatheronline{year}hcm.csv")
    data["Day"] = data["Date"].apply(split_d1)
    data["Month"] = data["Date"].apply(split_m1)
    data["year"] = data["Date"].apply(split_y1)
    data = data.drop(columns=["Date","Sunrise", "Sunset", "Moonrise", "Moonset","Icon","Province"])
    return data
    
    
```

```python
def parse_data_sung(year):
    data = process_data_freeme_sung(year)
    data1 = process_data_world_sung(year)
    merged_df = pd.merge(data1 , data , on=["Time","Day","Month","year"], how='inner')
    merged_df.drop_duplicates(subset=["Time","Day","Month","year"], keep='first', inplace=True, ignore_index=True)
    return merged_df
```

```python
years = [2017,2018,2019]
first = True
for year in years:
    if first == True:
        merged_df_sung = parse_data_sung(year)
        first = False
    else:
        merged_df_sung = merged_df_sung.append(parse_data_sung(year), ignore_index=True)
    print(len(merged_df_sung))
```

```python
def convert_wind_speed(x):
    if "°" in x:
        return x.split("°")[1]
    else:
        return x


def convert_wind_dir(x):
    if "°" in x:
        return x.split("°")[0] + "°"
    else:
        return x
    

def parse_data(year):
    data = pd.read_csv(f"../data/weather_data_[{year}].csv")
    data1 = pd.read_csv(f"../data/weather_data_[{year}]_1.csv")
    data = data.rename(columns={"Thời gian": "Time",'date':"Day","month":"Month"})
    data["Hướng gió"] = data["Gió"].apply(convert_wind_dir)
    data["Vận tốc gió"] = data["Gió"].apply(convert_wind_speed)
    data = data.drop(columns = ["Gió"])
    merged_df = pd.merge(data , data1 , on=['Time',"Day","Month"], how='inner')
    merged_df.drop_duplicates(subset=['Time',"Day","Month"], keep='first', inplace=True, ignore_index=True)
    return merged_df
```
